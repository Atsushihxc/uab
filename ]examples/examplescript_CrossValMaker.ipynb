### Create xval fold numbers by a list of file names
1. The filelist does not necessarily to be generated by uab_collectionFunctions
2. A file list generated by my own functions in the past is used in this example

```python
# define directories
parentDir = r'/media/ei-edl01/data/remote_sensing_data'
fileList = r'/media/ei-edl01/data/remote_sensing_data/collections/bohao_inria_train/bohao_inria_train.collect'
```

Take a look at file list:

```python
with open(fileList, 'r') as file:
    f = file.readlines()
for file in f:
    print(file)
```

This file list has files from 5 cities (austin, chicago, kitsap, tyrol-w and vienna) with tile numbers from 

Do xval by city first, we are expected to see a the folds idx range from 0 to 4

```python
import uabCrossValMaker
idx = uabCrossValMaker.uabUtilGetFolds(parentDir, fileList, 'city')
for id, file_name in zip(idx, f):
    print('{}: {}'.format(id, file_name.split(' ')[0].split('/')[-1]))
```
